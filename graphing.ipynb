```python
import json
import numpy as np
import matplotlib.pyplot as plt

with open("model_metrics.json", "r") as f:
    data = json.load(f)
```

```python
labels = [f"Patient {i+1}" for i in range(len(data["Text-Classification"]["f1_per_question"]))]
text_classification = data["Text-Classification"]["f1_per_question"]
openai = data["OpenAI"]["f1_per_question"]
semantic_similarity = data["Semantic-Similarity"]["f1_per_question"]
```

```python
x = np.arange(len(labels))
width = 0.2  

fig, ax = plt.subplots(figsize=(8, 5))
bars1 = ax.bar(x - width, text_classification, width, label="Text-Classification", color="#8ecae6")
bars2 = ax.bar(x, openai, width, label="OpenAI", color="#219ebc")
bars3 = ax.bar(x + width, semantic_similarity, width, label="Semantic-Similarity", color="#fb8500")

ax.set_xlabel("Patients")
ax.set_ylabel("F1 Score")
ax.set_title("F1 Score per Patient for Each Model")
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()
plt.ylim(0, 1.1)  
plt.show()
```
